```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint
from keras.utils import np_utils
import pandas as pd
```

```python
# Load the data
file = open('data/bandnames.txt')
# file = pd.read_csv('data/startups.csv')
# file.fillna('', inplace=True)
# data = file.name.values
# data.shape

# raw = ''.join(list(data))
data = [line.split('\n')[0] for line in file.readlines()]
data = [line.split('[')[0] for line in data]
raw = open('data/bandnames.txt').read()
```

Create mapping for each letter

```python
chars = sorted(list(set(raw)))
char_to_int = dict((c, i) for i, c in enumerate(chars))
char_to_int['NULL'] = -1
```

```python
L = []
for n, line in enumerate(data):
    L.append(len(line))
```

```python
n_chars = len(char_to_int)
mean_l = np.mean(L)
print('--- Dataset loaded ---')
print('{0} names in bandnames.text'.format(len(data)))
print('{0} unique letters mapped'.format(n_chars))
print('{0} average length'.format(mean_l))
print('------')
```

Map the sets to numbers

```python
dataX = np.ones((len(data),n_chars))*-1
dataY = np.ones((len(data),n_chars))*-1
for i, line in enumerate(data):
#     dataX.append([char_to_int.get(i,i) for i in line])
#     dataY.append([char_to_int.get(i,i) for i in line])
    ii = int(L[i]/2)
    dataX[i,:ii] = [char_to_int.get(n,n) for n in line]
    dataY[i,:] = [char_to_int.get(n,n) for n in line]
```

Define the model

```python
# reshape X to be [samples, time steps, features]
X = np.reshape(dataX, (dataX.shape[0], n_chars, 1))
# normalize
# X = X / float(n_vocab)
# one hot encode the output variable
# y = np_utils.to_categorical(dataY)
max_int = max(char_to_int.values())

y = dataY/n_chars
X = X/n_chars
```

```python
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
# Larger LSTM Network to Generate Text for Alice in Wonderland
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint
from keras.utils import np_utils
# load ascii text and covert to lowercase
# filename = "wonderland.txt"
# raw_text = open(filename).read()
raw_text = raw.lower()
# create mapping of unique chars to integers
chars = sorted(list(set(raw_text)))
char_to_int = dict((c, i) for i, c in enumerate(chars))
# summarize the loaded data
n_chars = len(raw_text)
n_vocab = len(chars)
print("Total Characters: ", n_chars)
print("Total Vocab: ", n_vocab)
# prepare the dataset of input to output pairs encoded as integers
seq_length = 100
dataX = []
dataY = []
for i in range(0, n_chars - seq_length, 1):
    seq_in = raw_text[i:i + seq_length]
    seq_out = raw_text[i + seq_length]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
n_patterns = len(dataX)
print("Total Patterns: ", n_patterns)
# reshape X to be [samples, time steps, features]
X = numpy.reshape(dataX, (n_patterns, seq_length, 1))
# normalize
X = X / float(n_vocab)
# one hot encode the output variable
y = np_utils.to_categorical(dataY)
# define the LSTM model
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
# define the checkpoint
filepath="weights-improvement-{epoch:02d}-{loss:.4f}-bigger.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
# fit the model
```

```python
model.fit(X[:147574,:,:], y[:147574,:], epochs=20, batch_size=128)
```
